```python
# Finding How many Numeric and Categorical features
numeric_features = df.select_dtypes(exclude=['object'])#.copy()
categorical_features = df.select_dtypes(include=['object'])#.copy()

print("Number of numerical features are: ", numeric_features.shape[1])
print("Number of categorical features are: ", categorical_features.shape[1])
```

```python
# Printing Numeric and Categorical features
numeric_features.columns, categorical_features.columns
```

```python
# converting categorical data into Numerical data
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for col in categorical_features.columns:
    df[col]=labelencoder.fit_transform(df[col])
df.head()
```

```python
df.tail()
```

```python
#seperating X and y
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
#data normalization
from sklearn import preprocessing
X =preprocessing.StandardScaler().fit_transform(X)
X
```

```python
#dividing into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=75)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
#prediction
y_pred = LR.predict(X_test)
print(LR.coef_)
print(LR.intercept_)
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```

```python
prediction_output =pd.DataFrame(data=[y_test,y_pred],index=['y_test','y_pred'])
prediction_output.transpose()
```
